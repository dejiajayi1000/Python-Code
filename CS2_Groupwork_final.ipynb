#Cleaning the Name column data#

```python
len(merge_df['Name'].unique())
```

```python
def standardize_name(name): 
    # Find the index of the first occurrence of a number 
    index = next((i for i, c in enumerate(name) if c.isdigit() and i>4 ), None) 
    # Remove everything after the first occurrence of a number 
    if index is not None: 
        name = name[:index] 
    # Remove any trailing whitespace characters 
    name = name.rstrip().title()
    characters_to_remove = ['.', ',','#', '|',':','%','!', '?', '/']
    words_to_remove = ['Barrel','L','but']
    for char in characters_to_remove:
        name = name.replace(char, '').replace('  ',' ').strip() 
 
    for char in words_to_remove:
        if name.endswith(char):
            name = name[:-len(char)]
    # Return the standardized name 
    return name 

# Apply the standardize_name function to each element of the column 
merge_df['Name'] = merge_df['Name'].apply(standardize_name) 

# Print the cleaned 'Name' column for randomly picked rows
print(merge_df.iloc[10000:20000,:]['Name'].unique())
```

Cleaning sale Amount data

```python
#Checking how many rows in the Sale Amount Column has null values
merge_df[merge_df['Sale Amount'].isnull()==True]
```

```python
#Taking the index values for every null - sale amount row and non_null - Retail price, Amount and Discount amount rows
index_sale_amt = merge_df[(merge_df['Retail Price'].isnull()==False) & (merge_df['Amount'].isnull()==False) & (merge_df['Discount Amount'].isnull()==False) & (merge_df['Sale Amount'].isnull()==True)].index
index_sale_amt
```

```python
#Calculating the sale amount based on Retail price, Amount and Discount amount rows
merge_df.loc[index_sale_amt,'Sale Amount']=round((merge_df.iloc[index_sale_amt]['Retail Price']*merge_df.iloc[index_sale_amt]['Amount'])-merge_df.iloc[index_sale_amt]['Discount Amount'],2)
merge_df.loc[index_sale_amt,'Sale Amount']
```

```python
#dropping the null values and negative values for sale amount
after_dropping_sa = merge_df.drop(merge_df[(merge_df['Sale Amount'].isnull()==True) | (merge_df['Sale Amount'] < 0)].index, inplace = False)
after_dropping_sa 
```

```python
#reviewing the null values for the rest of the columns in the dataframe after dropping missing sale amount rows.
after_dropping_sa.isnull().sum()
```

#Discount Amount & Discount Percentage#

```python
#Checking if the missing rows of both the 'Discount Amount' and 'Discount Percentage' are same
after_dropping_sa[(after_dropping_sa['Discount Amount'].isnull()==True) & (after_dropping_sa['Discount Percentage'].isnull()==True)].shape[0] == (after_dropping_sa['Discount Amount'].isnull().sum())
```

```python
#Selecting the indices with missing values in the 'Discount Amount' and 'Discount Percentage' columns
missing_discount_index = after_dropping_sa[after_dropping_sa['Discount Amount'].isnull()==True].index
missing_discount_index
```

```python
#Calculating the 'Discount Amount' for all null values rows and replacing them with the formula.
total_rp = (after_dropping_sa.loc[missing_discount_index,'Amount']*after_dropping_sa.loc[missing_discount_index,'Retail Price'])
sale_amount = after_dropping_sa.loc[missing_discount_index,'Sale Amount']

after_dropping_sa.loc[missing_discount_index,'Discount Amount'] = total_rp - sale_amount
```

```python
#Calculating the 'Discount Percentage' for all null values rows and replacing them with the formula.
discount_amount = after_dropping_sa.loc[missing_discount_index,'Discount Amount']
total_rp = after_dropping_sa.loc[missing_discount_index,'Retail Price']*after_dropping_sa.loc[missing_discount_index,'Amount']

after_dropping_sa.loc[missing_discount_index,'Discount Percentage'] = round((discount_amount / total_rp) * 100,2)
```

```python
#Check if there are any null values after calculating the 'Discount Amount' and 'Discount Percentage'.
after_dropping_sa['Discount Amount'].isnull().sum()
after_dropping_sa['Discount Percentage'].isnull().sum()
```

```python
#Setting the 'Discount Amount' and 'Discount Percentage' to zero when the values are calculated as negative.
mask = after_dropping_sa['Discount Percentage']<0
after_dropping_sa.loc[mask,'Discount Percentage'] = 0
after_dropping_sa.loc[mask,'Discount Amount'] = 0
```

#Visualization#

**Bar Graph**
It is important to understand the business revenue before making a business decision. it allows us to analyse how important it is to make a decision and how quickly we need to do it. The purpose of this visualization is to provide an overview of the sales over the past 3 years. The viewers can easily see the peaks and dips in the sales over the years.


```python
# Grouping the data by year and sum the sales amount
df_grouped = df.groupby(df['Date'].dt.year)['Sale Amount'].sum()
import matplotlib.pyplot as plt
# Creating a figure with two subplots
fig, ax1 = plt.subplots(figsize=(12, 7))

# Creating a bar chart on the first subplot
ax1.bar(df_grouped.index, df_grouped.values, color='C0')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Sales Amount (in 10 million)')
ax1.tick_params(axis='y')

# Setting the tick locations on the x-axis to be the same as the index of df_grouped
ax1.set_xticks(df_grouped.index)


for i, v in enumerate(df_grouped.values):
    ax1.annotate(str(v), xy=(i, v), ha='center', va='bottom')

# Create a second subplot and plot the trend on it
ax2 = ax1.twinx()
ax2.plot(df_grouped.index, df_grouped.values, color='C3')
ax2.set_ylabel('Trend')
ax2.tick_params(axis='y')

# Setting the title and legend
ax1.set_title('Total Sales and Trend by Year')
ax2.legend(['Trend'], loc='upper right')

# Showing the chart
plt.show()
```

 

In the bar chart we can see that the sales have increased from 2020 to 2021 but has seen a drastic decrease in 2022. This shows us the urgent need to change the strategy and understand what we can do to develop a better stratergy based on the previous years data
