```python
import random
def roll_dice(number_of_dice,number_of_sides):
    print(f'Rolling {number_of_dice} dice with {number_of_sides} sides each...')
    total= 0
    highest = 0
    for x in range (1,number_of_dice +1):
        value = random.randint(1,number_of_sides)
        if value > highest:
            highest = value
        total = total + value
        print(f'Rolled {value}')
    print(f'The total is {total}')
    print(f'The highest number rolled is {highest}')
    
```
